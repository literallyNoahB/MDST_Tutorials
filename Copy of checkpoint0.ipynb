```python
# optional challenge - fizzbuzz
# you will need to use both iteration and control flow 
# go through all numbers from 1 to 100 in order
# if the number is a multiple of 3, print fizz
# if the number is a multiple of 5, print buzz
# if the number is a multiple of 3 and 5, print fizzbuzz and NOTHING ELSE
# if the number is neither a multiple of 3 nor a multiple of 5, print the number
for i in range(1,101):
  if i % 3 == 0 and i % 5 == 0:
    print("fizzbuzz")
  elif i % 3 == 0:
    print("fizz")
  elif i % 5 == 0:
    print("buzz")
  else:
    print(i)
```

```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial

# the tutorial is not working, I'm just going to circumvent the problem
import pandas as pd
from google.colab import drive

# mount drive
drive.mount('/content/gdrive')

# read imn csv
df = pd.read_csv("gdrive/My Drive/Clubs/Data Science Team/2022-23/Copy of starbucks.csv")
df.head()
```

```python
# output the calories, sugars, and protein columns only of every 40th row. 
x, y = df.shape
rows = list(range(40,x,40))
df.loc[rows, ["calories", "sugars", "protein"]]
```

```python
# select all rows with more than and including 400 calories
keyRows = df.loc[df["calories"] >= 400].index
keyRows
```

```python
# select all rows whose vitamin c content is higher than the iron content
keyRows = df.loc[df["vitamin c"] > df["iron"]].index
keyRows
```

```python
# create a new column containing the caffeine per calories of each drink
CafCal = df.loc[:, "caffeine"] / df.loc[:, "calories"]
df["Caffeine per Calories"] = CafCal
df.head()
```

```python
# what is the average calorie across all items?
avgCal = df.loc[:, "calories"].mean()
avgCal
```

```python
# how many different categories of beverages are there?
beverages = df.loc[:, "beverage_category"].unique()
print(len(beverages))
```

```python
# what is the average # calories for each beverage category?
avgCalories = {i: df.loc[df["beverage_category"] == i].loc[:, "calories"].mean() for i in beverages}
avgCalories
```

```python
# plot the distribution of the number of calories in drinks with a histogram
import matplotlib.pyplot as plt
df.loc[:, "calories"].hist()
plt.title("Histogram of Number of Calories in Drinks")
plt.show()
```

```python
# plot calories against total fat with a scatterplot
fig, ax = plt.subplots(figsize=(9,6))
plt.scatter(df.loc[:, "total fat"], df.loc[:, "calories"], c='g')
plt.xlabel("Total Fat")
plt.ylabel("Calories")
plt.show()
```
